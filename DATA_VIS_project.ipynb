```python
# import plotly.graph_objects as go
# from plotly.subplots import make_subplots
from os import listdir
import pandas as pd
import plotly.express as px
fh=["2011-12","2015-16","2018-19","2019-20","2020-21"]
gn=listdir("/home/kar/Desktop/jupyter/ZP/")
gn=[gn[i].split(".")[0] for i in [3,9,25,0]]
g1 = pd.read_excel("~/Desktop/jupyter/Chhatisgarh-1.xlsx",
                       sheet_name="Sheet6",engine='openpyxl').transpose()
g1.columns=g1.loc["Item"]
g3=g1.drop(index="Item")
for kl in gn:
    g = pd.read_excel("~/Desktop/jupyter/ZP/"+kl+".xlsx",
                    sheet_name="Sheet5",engine='openpyxl').transpose()
    g.columns=g.loc["Item"]
    g4=g.drop(index="Item")
    g4.insert(0,'sector',g3["sector"])
    g4.insert(0,'subsec',g3["subsec"])
    g4=g4.dropna(axis=0)
    for time in fh:
        l=[]
        for x in range(len(g4.index)):
            if g4.index[x] == g4["subsec"][x]:
                l.append(None)
            else:
                l.append(g4.index[x])
        try:
            fo=px.sunburst(g4,path=[g4["sector"],g4["subsec"],l],
                   values=g4[time],
                      color_discrete_sequence=px.colors.qualitative.D3
                      )
            fo.update_traces(
                textinfo="label+percent entry",
            )
            fo.update_layout(
            height=800,
                title_text="Sector Tree in "+kl+" for "+time,
                title_font_family="Times New Roman",
                title_font_size=23,
                title_x=0.5
            )
            # fo.write_html(file="/home/kar/Downloads/graphs/"+kl+"_aseccompst"+time+".html")
            fo.show()
        except :
            print("not next ",kl,time)
        break
```

```python
from os import listdir
import pandas as pd
# import plotly.express as px
# fh=["2011-12","2015-16","2018-19","2019-20","2020-21"]
gn=listdir("/home/kar/Desktop/jupyter/ZP/")
gn=[gn[i].split(".")[0] for i in [4,9,10]]
import plotly.graph_objects as go
from plotly.subplots import make_subplots
# import pandas as pd
for kl in gn:
    gf = pd.read_excel("~/Desktop/jupyter/ZP/"+kl+".xlsx"
                       ,sheet_name="Sheet5",engine='openpyxl')
    label1=gf.keys()[[1,6]]
    label2=gf.keys()[[8,9,10]]
    label3=gf.keys()[[12,15,23,24,25,26]]
    a={0:"2011-12",4:"2015-16",7:"2018-19",8:"2019-20",9:"2020-21"}
    for i in a.keys():
        try:
            l1=[gf.loc[i,x] for x in label1]
            l2=[gf.loc[i,x] for x in label2]
            l3=[gf.loc[i,x] for x in label3]
            fig = make_subplots(rows=2,
                                cols=2,
                                specs=[[{'type':'pie',"colspan":2},None],
                                        [{'type':'pie'},{'type':'pie'}]]
                                ,horizontal_spacing=0,
                                vertical_spacing=0
                               )
            fig.add_trace(go.Pie(
                legendgroup="Primary",
                                 labels=label1,
                                 values=l1,
                                 name="Primary",
                                ),
                          row=1,col=1)
            fig.add_trace(go.Pie(
                legendgroup="Secondary",
                                 values=l2,
                                 labels=label2,
                                 name="Secondary",
                                ),

                          row=2,col=2)
            fig.add_trace(go.Pie(
                legendgroup="Tertiary",
                                 values=l3,
                                 labels=label3,
                                 name="Tertiary"
                                ),
                          row=2, col=1
                         )
            fig.update_traces(hole=0.4,hoverinfo="label+name+value")
            fig.update_layout(
                legend=dict(
                            font_family="Times New Roman",
                            font_size=13,
                            borderwidth=1
                           ),
                margin=dict(b=0,l=0,r=0),
                width=1000,
                height=600,
                title_font_family="Times New Roman",
                title_font_size=23,
                title_x=0.5,
                title_text="SubSector in "+kl+" Contributing for GSDP-"+a[i],
                annotations=[dict(text='Primary', x=0.5, y=0.78, font_size=15, showarrow=False),
                             dict(text='Tertiary', x=0.20, y=0.23, font_size=15, showarrow=False),
                             dict(text='Secondary', x=0.82, y=0.23, font_size=14, showarrow=False)])
            fig.write_html(file="/home/kar/Downloads/graphs/"+kl+"_subcompst"+a[i]+".html")
        except:
            print(kl," ",a[i])
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
from os import listdir
gn=listdir("/home/kar/Desktop/jupyter/ZP/")
gn=[x.split(".")[0] for x in gn]
for kl in gn:
    g = pd.read_excel("~/Desktop/jupyter/ZP/"+kl+".xlsx",
                       sheet_name="Sheet5",engine='openpyxl').transpose()
    # import plotly.graph_objects as go
    # from plotly.subplots import make_subplots
    g.columns=g.loc["Item"]
    g=g.drop(index="Item")
    g=g.drop(index=g.index[26:33])
    g=g.drop(index=g.index[10])
    g=g.drop(index=g.index[6])
    g=g.dropna()
    l=len(g.keys())
    a1=[]
    for i in range(len(g.index)):
        x=g.iloc[i]
        d=sorted([x[j+1]-x[j] for j in range(l-1)])
        a1.append(sum(d)/l)
    b=list(enumerate(a1))
    s=[x[1] for x in sorted(b,key=lambda k:k[1])[-5:]]# last  top 5
    s1=[x[0] for x in sorted(b,key=lambda k:k[1])[-5:]]
    if any(x<0 for x in s) == False:
        maxx=px.pie(values=s,names=[g.index[x] for x in s1])
        maxx.update_layout( title="Top 5 Sectors in Growth Rate of "+kl,
                      title_font_family="Times New Roman",
                        title_font_size=23,
                        title_x=0.5,
                      )
#         print("not there")
    else:
        x1=np.where(np.array(s) < 0,"-VE","+VE")
        maxx=px.bar(y=s,x=[g.index[x] for x in s1],
        color=x1,
        color_discrete_sequence=px.colors.qualitative.Light24,
       labels={"x":"Factors","y":"Growth"}
                   )
        maxx.update_layout( title="Top 5 Sectors in Growth Rate of "+kl,
                        title_font_family="Times New Roman",
                        title_font_size=23,
                        title_x=0.5,
                      )
#         print("is")
    h=[x[1] for x in sorted(b,key=lambda k:k[1])[0:5]]# first 5 bottom
    h1=[x[0] for x in sorted(b,key=lambda k:k[1])[0:5]]
    if any(x<0 for x in h) == False:
        minn=px.pie(values=h,names=[g.index[x] for x in h1])
        minn.update_layout( title="Bottom 5 Sectors in Growth Rate of "+kl,
                        title_font_family="Times New Roman",
                        title_font_size=23,
                        title_x=0.5,
                      )
#         print("not there ")
    else:
        x1=np.where(np.array(h) < 0,"-VE","+VE")
        minn=px.bar(y=h,x=[g.index[x] for x in h1],
        color=x1,
        color_discrete_sequence=px.colors.qualitative.Light24,
       labels={"x":"Factors","y":"Growth"}
                   )
        minn.update_layout( title="Bottom 5 Sectors in Growth Rate of "+kl,
                        title_font_family="Times New Roman",
                        title_font_size=23,
                        title_x=0.5,
                      )
#         print("is")
    minn.write_html(file="/home/kar/Downloads/graphs/"+kl+"_grb.html")
    maxx.write_html(file="/home/kar/Downloads/graphs/"+kl+"_grt.html")
```
